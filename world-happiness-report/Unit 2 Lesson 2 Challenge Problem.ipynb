```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv('2016.csv')
print (df)
```

```python
df.columns
```

```python
df.head()
```

```python
sns.swarmplot(x='Region', y='Happiness Score', data=df)
plt.xticks(rotation=85)
```

```python
#We can see that North Western Europe has the greatest amount of happy 
#countries.Even though they have a spread, it never dips into 
#minimum score like Sub-Saharan Africa
```

```python
#This graph shows that first world countries are happier compared to 
#third world countries like Africa and South Asia.
#However, this graph isn't really descriptive but shows the spread of the 
#regional happiness.
```

```python
w_europe = df[df.Region=='Western Europe']
ec_europe = df[df.Region=='Central and Eastern Europe']
europe = pd.concat([w_europe,ec_europe],axis=0)
europe.head()
```

```python
f, axes = plt.subplots(4, 1, figsize=(12, 12))
axes = axes.flatten()
compareCols = ['Happiness Score','Economy (GDP per Capita)','Generosity','Health (Life Expectancy)']
for i in range(len(compareCols)):
    col = compareCols[i]
    axi = axes[i]
    sns.distplot(w_europe[col],color='blue' , label='West', ax=axi)
    sns.distplot(ec_europe[col],color='green', label='Central/East',ax=axi)
    axi.legend()
```

#We see that W. Europe has a higher happiness score compared to C/E Europes score.
#We could make the inference that because W. Europe is happier, 
#W. Europe will also have a higher GDP, longer life expectancy which both happen to be true.
# I assumed that a happier region would also be more generous
#however, it seems like Central/East Europe, although not as happy
#have the capacity to be more generous than their neighbors.

```python
gen = df['Generosity']
happy = df['Happiness Score']
plt.scatter(gen, happy)
plt.xlabel('Generosity')
plt.ylabel('Happiness Score')
```

```python
#It seems like the world population follows the trend of W.Europe and C/E. Europe
#There is one outlier that is extremely generous but also not very happy.
#Lastly, seems like whether happy or unhappy, 
#the majority have the generosity of 0.2
```

```python
corruption = df['Trust (Government Corruption)']
plt.scatter(corruption, happy)
plt.xlabel('Corruption')
plt.ylabel('Happiness Score')
```

```python
# General morality tells us that a corrupt government would make its citizens
#unhappy either because of oppression or embezzlement or other reasons.
#It seems like the happiest countries still has a corruption level around .4 to .5.
#Only one outlier appeals to the initial hypothesis that a corrupt gov. 
#would yield a low happiness score.
#Generally speaking though, the majority of countries whether happy or 
#unhappy experience a level of corruption up to 0.2.
```
